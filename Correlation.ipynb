```python
import pandas as pd
import numpy as np
import random
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
y = [2158.70, 1768.15, 2316.00, 2061.30, 2207.50, 1708.30, 1784.70, 2575.00,2357.90, 2056.70]
x = [15.50, 23.75, 8.00, 17.00, 5.50, 19.00, 24.00, 2.50, 7.50, 11.00]

```

```python
corr = np.corrcoef(x,y)

print(" corr: r = ",corr)
```

correlation coefficient between shear stength and age of propellant is r = - 0.91 < 0 
stong negative correlation between shear strength and age of propellant.
Implies that increasing in age of propellant the shear stength decrease.


```python
plt.figure(dpi = 100)
plt.scatter(x,y,color = 'green', label = 'Correlation')
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Scatter plot : X vs Y ")
plt.legend()
plt.show()
```

Conclsion : From the above scatter plot it can be clearly visualize that there seems to be a negative correlation between random variable x and y
