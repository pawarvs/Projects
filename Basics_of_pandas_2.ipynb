#### Merge queries

```python
data4 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
```

```python
data5  = {'emp_id' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df4  = pd.DataFrame(data4)
df4
```

```python
df5 = pd.DataFrame(data5)
df5
```

```python
pd.merge(df4,df5)
```

```python
pd.merge(df4,df5, how = 'left')
```

```python
pd.merge(df4,df5, how = 'right',on = 'emp_id')
```

```python
data6 = {'emp_id1' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
```

```python
data7  = {'emp_id2' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df6 = pd.DataFrame(data6)
```

```python
df7 = pd.DataFrame(data7)
```

```python
# with no common columns
pd.merge(df6,df7,left_on = 'emp_id1',right_on = 'emp_id2',how = 'inner')
```

```python
data8 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
data9  = {'emp_id' : [101,102,104,105,123],
          'salary' : [234, 345,200,400,532],
          'house_nu' : [345,345,345,56,494]
    
}
```

```python
df8 = pd.DataFrame(data8)
df9 = pd.DataFrame(data9)
df9
```

```python
df8
```

```python
# Meger operation performed based on columns.
pd.merge(df8,df9,on=['emp_id','salary'])
```

```python
df_sales = pd.read_excel("""D:\Power Bi\Datasets\Sample - Superstore.xlsx""")
```

```python
df_sales.head()
```

```python
df_sales.dtypes
```

```python
def profit_flag(a) :
    
    if a >= 0 :
        return 'positive'
    else :
        return 'negative'
```

```python
profit_flag(0)
```

```python
profit_flag(48)
```

```python
profit_flag(-20)
```

```python
df_sales['profit_flag'] = df_sales['Profit'].apply(profit_flag)
df_sales.head()
```

```python
df_sales['len_cust_name'] = df_sales['Customer Name'].apply(len)
df_sales.head()
```

```python
df_sales['Quantity'].unique()
```

```python
df_sales['Profit'].unique()
```

```python
def flag_profit(a) :
    if a < 50 :
        return 'Low'
    elif a > 51 and a < 65 :
        return 'Medium'
    else :
        return 'High'
```

```python
df_sales['flag_profit'] = df_sales['Profit'].apply(flag_profit)
df_sales.head()
```

```python
df_sales['Added_quantity'] = df_sales['Quantity'].apply(lambda a : a + 15)
```
